키오스크 만들기

```python
class Kiosk:
    def __init__(self):
        self.menu = ['Americano','Cafe_latte','Cappuccino','Cold_Brew','Green_Tea','Red_tea','Milk_Tea']
        self.price = [3000,4000,4000,4500,3000,3000,4000]
        self.order_menu = []
        self.order_price = []

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ':', self.price[i], '원')

    def menu_select(self):
        n = int(input("메뉴의 번호를 입력하세요: "))
        self.price_sum = self.price[n - 1]
        print(n, self.menu[n - 1], ":", self.price[n - 1], "원")

        t = 0
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp ="ICE"
            else:
                print("다시 입력하세요.\n")
        self.order_price.append(self.price[n - 1])
        self.order_menu.append(self.temp + " " + self.menu[n - 1])

        print('추가 주문 메뉴',self.temp,self.menu[n - 1], ":",self.price[n - 1],"원")

        while n != 0:
            n = int(input("추가 주문은 메뉴 번호를, 지불은 0을 누르세요: "))

            if n > 0 and n < len(self.menu) + 1:
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("다시 입력하세요.\n")

                self.order_price.append(self.price[n - 1])
                self.order_menu.append(self.temp + " " + self.menu[n - 1])
                self.price_sum += self.price[n - 1]
                print('추가 주문 음료',self.temp,self.menu[n-1],':',self.price[n-1],'원\n','합계 :',self.price_sum,'원')
            else:
                if n == 0:
                    print('주문이 완료되었습니다.')
                    print(self.order_menu, self.order_price)
                else:
                    print('없는 메뉴입니다.')
    def pay(self):
        print(f"총 합계 금액은 {self.price_sum}원 입니다.")
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(
                input("현금 결제는 1번, 카드 결제는 2번을 입력해주세요.: "))
            if self.pay_method == 1:
                print("직원을 호출하겠습니다.")
            elif self.pay_method == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
            else:
                print("다시 결제를 시도해 주세요.")

    def table(self):
        print("\n===== 주문표 =====")
        print("-----------------------------")
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], "|", self.order_price[i], "원")
        print("-----------------------------")
        print("총 결제 금액 :", self.price_sum, "원")

a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
