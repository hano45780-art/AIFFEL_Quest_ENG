# Q1 은행계좌 만들기

```python
import random

class Account:
    # 생성된 총 계좌 개수를 관리하는 클래스 변수
    account_count = 0

    def __init__(self, owner, balance):
        # 인스턴스 속성 초기화
        self.bank_name = "SC은행"
        self.owner = owner
        self.balance = balance
        self._deposit_count = 0  # 이자 지급용 입금 횟수 누적

        # 입출금 기록용 리스트 초기화
        self._deposit_records = []
        self._withdraw_records = []

        # 3자리-2자리-6자리 형태의 랜덤 계좌번호 생성
        self.account_number = (
            f"{random.randint(100, 999)}-"
            f"{random.randint(10, 99)}-"
            f"{random.randint(100000, 999999)}"
        )

        # 계좌 개수 클래스 변수 1 증가
        Account.account_count += 1

    @classmethod
    def get_account_num(cls):
        # 생성된 총 계좌 개수 반환
        return cls.account_count

    def deposit(self, amount):
        # 1원 이상 입금 검증
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return

        self.balance += amount
        self._deposit_count += 1
        self._deposit_records.append(amount)  # 입금 내역 기록

        # 5회 입금 시 잔고 기준 1% 이자 지급
        if self._deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest

    def withdraw(self, amount):
        # 잔고 이하 금액 출금 검증
        if amount > self.balance:
            print("출금은 계좌의 잔고 이상으로 할 수 없습니다.")
            return

        self.balance -= amount
        self._withdraw_records.append(amount)  # 출금 내역 기록

    def display_info(self):
        # 계좌 정보 종합 출력 (잔고는 세자리마다 쉼표 출력)
        print(
            f"은행이름: {self.bank_name}, "
            f"예금주: {self.owner}, "
            f"계좌번호: {self.account_number}, "
            f"잔고: {self.balance:,}원"
        )

    def deposit_history(self):
        # 입금 내역 출력
        if not self._deposit_records:
            print("입금 내역이 없습니다.")
            return
        print(f"[{self.owner}] 입금 내역: {self._deposit_records}")

    def withdraw_history(self):
        # 출금 내역 출력
        if not self._withdraw_records:
            print("출금 내역이 없습니다.")
            return
        print(f"[{self.owner}] 출금 내역: {self._withdraw_records}")


# ==================== 실전 구동 영역 ====================
if __name__ == "__main__":
    # 4개의 계좌 객체 생성 및 리스트 관리
    accounts = [
        Account("파이썬", 10_000),
        Account("자바", 500_000),
        Account("C++", 1_500_000),
        Account("자바스크립트", 800_000),
    ]

    # 총 생성된 계좌 수 확인
    print(f"생성된 계좌 수: {Account.get_account_num()}")
    print()

    # 기능 테스트용 입출금 연쇄 가동 (5회 입금으로 이자 트리거 작동)
    accounts[0].deposit(5_000)
    accounts[0].deposit(3_000)
    accounts[0].deposit(2_000)
    accounts[0].deposit(1_000)
    accounts[0].deposit(4_000)
    accounts[0].withdraw(2_000)

    # 잔고 100만 원 초과 달성을 위한 추가 입금
    accounts[1].deposit(600_000)

    print("=== 전체 계좌 정보 ===")
    for account in accounts:
        account.display_info()

    print()
    print("=== 잔고 100만원 이상 우수 고객 필터링 ===")
    for account in accounts:
        if account.balance >= 1_000_000:
            account.display_info()

    print()
    print("=== 고객별 입출금 내역 ===")
    for account in accounts:
        account.deposit_history()
        account.withdraw_history()
```

# Q2 간단한 자동사냥 RPG 만들기

```python
import random

# =========================
# Q1. Character 클래스
# =========================
class Character:
    def __init__(self, name, level, hp, attack, defense):
        """
        모든 캐릭터의 기본 속성
        """
        self.name = name
        self.level = level
        self.hp = hp
        self.attack_power = attack
        self.defense = defense

    # 현재 살아있는지 확인
    def is_alive(self):
        """
        체력이 0 이상이면 True
        """
        return self.hp > 0

    # 데미지를 받는 함수
    def take_damage(self, damage):
        """
        (받은 데미지 - 방어력)만큼 체력 감소
        단, 방어력이 더 크면 데미지 0 처리
        """

        real_damage = damage - self.defense

        if real_damage < 0:
            real_damage = 0  # 방어력이 더 크면 데미지 없음

        self.hp -= real_damage

        print(f"{self.name}이(가) {real_damage} 피해를 입었습니다. (HP: {self.hp})")

    # 공격 함수
    def attack_target(self, target):
        """
        1 ~ 공격력 사이 랜덤 데미지로 공격
        """

        damage = random.randint(1, self.attack_power)
        print(f"{self.name} → {target.name} 공격! (데미지: {damage})")
        target.take_damage(damage)


# =========================
# Q2. Player 클래스
# =========================
class Player(Character):
    def __init__(self, name):
        """
        플레이어 초기값 고정:
        레벨 1 / HP 100 / 공격 25 / 방어 5
        """
        super().__init__(name, 1, 100, 25, 5)
        self.exp = 0  # 경험치

    # 경험치 획득
    def gain_experience(self, exp):
        self.exp += exp
        print(f"{self.name} 경험치 +{exp} (현재 EXP: {self.exp})")

    # 레벨업
    def level_up(self):
        """
        EXP 50 이상이면 레벨업
        레벨 +1
        공격력 +10
        방어력 +5
        """
        while self.exp >= 50:
            self.exp -= 50
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            self.hp = 100  # 레벨업 시 체력 회복

            print(f"🎉 레벨업! 현재 레벨: {self.level}")


# =========================
# Q2. Monster 클래스
# =========================
class Monster(Character):
    def __init__(self, level):
        """
        몬스터는 레벨 기반 스탯 랜덤 생성
        """

        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(f"Monster(Lv{level})", level, hp, attack, defense)


# =========================
# Q3. battle 함수
# =========================
def battle(player, monster):
    """
    플레이어 vs 몬스터 자동 전투
    """

    print("\n⚔️ 전투 시작!")
    print(f"{player.name} vs {monster.name}\n")

    # 둘 중 하나 죽을 때까지 반복
    while player.is_alive() and monster.is_alive():

        # 플레이어 공격
        player.attack_target(monster)

        # 몬스터가 죽었는지 체크
        if not monster.is_alive():
            break

        # 몬스터 공격
        monster.attack_target(player)

    # =========================
    # 전투 결과 처리
    # =========================
    if player.is_alive():
        print("\n🎉 전투 승리!")

        # 경험치 획득
        exp_gain = monster.level * 20
        player.gain_experience(exp_gain)

        # 레벨업 체크
        player.level_up()

    else:
        print("\n💀 전투 패배..")


# =========================
# 게임 실행 부분
# =========================

name = input("플레이어 이름을 입력하세요: ")
player = Player(name)

# 몬스터 3마리 자동 생성 후 전투
for i in range(1, 4):
    print(f"\n========== {i}번째 몬스터 등장 ==========")
    monster = Monster(level=i)
    battle(player, monster)

    if not player.is_alive():
        print("\n게임 오버!")
        break
```
