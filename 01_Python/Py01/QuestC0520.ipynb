Q1-2 회문 분류기

```python
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

def is_palindrome(s: str) -> bool:
    return s == s[::-1]

def palindrome_classifier(words: list[str]):
    result = {"회문O" : [], "회문X" : []}
    for word in words:
        if is_palindrome(word):
            result["회문O"].append(word)
        else:
            result["회문X"].append(word)
    total = len(words)
    count = len(result["회문O"])
    ratio = count / total if total > 0 else 0.0
    return result, count, ratio
result, count, ratio = palindrome_classifier(targets)
print(f"회문 O: {result['회문O']}, 회문 X: {result['회문X']}, {count}, {ratio}")
```

Q2-2 성적 분석기

```python
scores = {
    "유원영": {"국어": 94, "영어": 91, "수학": 89},
    "설지민": {"국어": 88, "영어": 96, "수학": 90},
    "신윤아": {"국어": 86, "영어": 88, "수학": 77},
    "장윤아": {"국어": 98, "영어": 76, "수학": 92}
}
def score_analysis(name: str):
    if name not in scores:
        print(f"{name} 학생의 성적 데이터가 없습니다.")
        return
    
    subjects = list(next(iter(scores.values())).keys())
    subject_avg = {subj: sum(scores[stu][subj] for stu in scores) / len(scores)
                   for subj in subjects}
    
    student_scores = scores[name]
    
    print(f"[{name} 성적 분석]")
    for subj, score in student_scores.items():
        diff = score - subject_avg[subj]
        sign = "+" if diff >= 0 else ""
        print(f"{subj}: {score}점, 학급 평균({subject_avg[subj]:.2f}점) 대비 {sign}{diff:.2f}")
    
    student_avg = sum(student_scores.values()) / len(student_scores)
    overall_avg = sum([score for subj_scores in scores.values() for score in subj_scores.values()]) / (len(scores) * len(subjects))
    diff_total = student_avg - overall_avg
    sign_total = "+" if diff_total >= 0 else ""
    print(f"\n전과목 평균: {student_avg:.2f}점, 학급 평균({overall_avg:.2f}점) 대비 {sign_total}{diff_total:.2f}")

score_analysis("설지민")
```
